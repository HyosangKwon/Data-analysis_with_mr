## `pandas` 함수 조합하기
### `query()`와 `[]` 조합하기
- 1반 학생의 영어 점수 추출

```python
# nclass가 1인 행만 추출한 다음 English 추출
exam.query('nclass == 1')['english']
```

```python
# 두줄로 나누기
exam_nclass1 = exam.query('nclass == 1')
exam_nclass1['english']
```

- 수학 점수가 50점 이상인 학생의 `id`와 `math` 변수 추출

```python
# math가 50 이상인 행만 추출한 다음 id, math 추출
exam.query('math >= 50')[['id','math']]
```

일부만 추출하기

```python
exam.query('math >= 50')[['id','math']].head(10)
```

### 가독성 있게 코드 줄 바꾸기

```python
exam.query('math >= 50')\
 [['id','math']] \
 .head(10)
```

## 파생변수 추가: `df.assign()`

```python
# total 변수 추가
exam.assign(total = exam['math'] + exam['english'] + exam['science'])
```

```python
exam.assign(
    total1 = exam['math'] + exam['english'] + exam['science'],
    mean = (exam['math'] + exam['english'] + exam['science'])/3)
```

```python
import numpy as np
exam.assign(test = np.where(exam['english'] >= 70, '통과', '불통'))
```

```python
# 영어점수가 70 점 이상인 1반 학생의 데이터 추출해서 '통과' 그렇지 않으면 '불통'

# exam.query('nclass == 1').assign(test1 = np.where(exam['english'] >= 70, '통과', '불통'))
# 추출전 후 데이터 프레임 사이즈 달라지므로 새로운 변수에 따로 저장한 뒤 assign 하기

exam_n1 = exam.query('nclass == 1')
exam_n1.assign(test1 = np.where(exam_n1['english'] >= 70, '통과', '불통'))
```

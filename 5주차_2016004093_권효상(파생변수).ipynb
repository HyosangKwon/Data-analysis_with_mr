# 파생변수 만들기

### mpg 통합 연비 변수 만들기

```python
mpg['total'] = (mpg['cty'] + mpg['hwy'])/2 # 통합 연비 변수 만들기
mpg.head()
```

### 파생변수 분석하기

```python
sum(mpg['total'])/len(mpg) # total 합계를 행 수로 나눔
```

```python
mpg['total'].mean() # 통합 연비 변수 평균
```

### 조건문을 활용한 파생변수 만들기

```python
mpg['total'].describe()
```

### 히스토그램 그리기

```python
mpg['total'].plot.hist()
```

```python
import numpy as np

# 연비 20 이상이면  pass, 그렇지 않으면 fail 부여
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')

mpg.head()
```

### 막대 그래프로 빈도 표현하기

```python
count_test = mpg['test'].value_counts() # 연비 합격 빈도표를 변수에 할당
count_test.plot.bar()
```

# 축 이름 회전하기

```python
 count_test.plot.bar(rot = 0) # 축 이름 수평으로 만들기
```

# 중첩 조건문 활용

```python
count_grade.plot.bar(rot = 0)
```

### 목록에 해당하는 행으로 변수 만들기
- | : 버티컬 바
    - '또는(or)'을 의미

```python
mpg['size'] = np.where((mpg['category'] == 'compact')|
                       (mpg['category'] == 'subcompact')|
                       (mpg['category'] == '2seater'),
                       'small', 'large' )
mpg['size'].value_counts()
```

```python
mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']), 'small', 'large')

mpg['size'].value_counts()
```

```python
mpg['size'].value_counts().plot.bar()
```

# 분석 도전

- `midwest.csv` 는 미국 동북중부 437개 지역의 인구통계 정보 담고 있음.

------------------------------------
>### `midwest.csv`를 불러와 데이터의 특징 파악

```python
midwest = pd.read_csv('/content/drive/MyDrive/Data/midwest.csv')
midwest.describe()
```

```python
# 복사
df_midwest = midwest.copy() # 복사본 만들기
```

>### `poptotal(전체인구)` 변수를 `total`로, `popasian(아시아인구)` 변수를 `asian`으로 수정

```python
# 변수명 변경
df_midwest = df_midwest.rename(columns = {'poptotal':'total'})
df_midwest = df_midwest.rename(columns = {'popasian':'asian'})

df_midwest.head()
```

>### `total`, `asian` 변수 이용해 `'전체 인구 대비 아시아 인구 백분율'` 파생변수 추가, 히스토그램 만들어 분포 살펴보기


```python
# 파생변수 만들기

df_midwest['ratio'] = (df_midwest['asian']/df_midwest['total'])*100
df_midwest['ratio']
```

```python
## 히스토그램
df_midwest['ratio'].plot.hist()
```

>### 아시아 인구 백분율 전체 평균 구하고 평균 초과시 `'large'`, 그 외에는 `'small'`을 부여한 파생변수 만들기

```python
# 파생변수 만들기

ratio_mean = df_midwest['ratio'].mean()
ratio_mean
```

```python
df_midwest['group'] = np.where(df_midwest['ratio']>=ratio_mean, 'large', 'small' )
df_midwest.head()
```

>### `'large'`와 `'small'`에 해당하는 지역이 얼마나 많은지 빈도표와 빈도 막대 그래프를 만들어 확인

```python
# 빈도표
df_midwest['group'].value_counts()
```

```python
# 빈도 막대 그래프
df_midwest['group'].value_counts().plot.bar()
```

### 자동차 제조 회사에 따라 도시 연비가 어떻게 다른지 확인 `'audi'`와 `'toyota'` 중 어느 `manufacturer(제조 회사)`의 `cty(도시연비)` 평균이 더 높은지 알아보기

```python
# 도요타
toyota = mpg['manufacturer'] == 'toyota'
mpg[toyota].head()
```

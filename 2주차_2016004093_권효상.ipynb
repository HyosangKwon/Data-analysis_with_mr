#예제1

5명의 학생의 점수의 평균을 출력하시오.

```python
scores = [100,90,95,85,80]
sum_scores = 0

for i in scores:
    sum_scores += i

avg_scores = sum_scores / len(scores)

print(avg_scores)
```

```python
list = [100,90,95,85,80]
print("평균 : ", sum(list)/len(list))
```

### 패키지 로드하기

```python
import seaborn # 그래프 관련 패키지

var = ['a','a','b','c','b','d','f','d','a']
var

seaborn.countplot(x = var) # 리스트 값을 x로
```

### 패키지 약어 활용하기

```python
import seaborn as sns
sns.countplot(x = var)
```

# seaborn의 titanic 데이터로 그래프 만들기

```python
df =  sns.load_dataset('titanic')
df
```

### 함수의 다양한 기능 이용

```python
sns.countplot(data = df, x = 'sex')
```

```python
# x축 class
sns.countplot(data = df, x = 'class')
```

--------------
데이터 분석을 할 때는 데이터의 항목에 대한 이해가 중요하다.

```python
# 각 class 별 alive 색깔로 표현

sns.countplot(data = df, x = 'class', hue = 'alive')
```

```python
# y축 class, alive별 색 표현 (그래프 가로로 그리기)

sns.countplot(data = df, y = 'class', hue = 'alive')
```

```python
# 함수에 대해 상세설명을 보고 싶을 때 '?' 사용
sns.countplot?
```
